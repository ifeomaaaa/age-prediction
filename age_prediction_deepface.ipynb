```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('/content/IMG_5647.jpg')
plt.imshow(img1[:,:,::-1])
plt.show()
```

```python
result = DeepFace.analyze(img1, actions = ['age'])
```

```python
print('Predicted age is : ', result['age'])
```
